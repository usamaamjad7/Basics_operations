```python

import torch
import numpy as np
```

```python
# Tensors can be created directly from data

data = [[1,2],[3,4]]
x_data = torch.tensor(data)
```

```python
# Tensors can be created from Numpy array.
np_array = np.array(data)
x_np = torch.from_numpy(np_array)
```

```python
# From another tensor
torch.rand(4)
torch.rand(2, 3)
```

```python
x_ones = torch.ones_like(x_data) # retains the properties of x_data
print(f"Ones Tensor: \n {x_ones} \n")

x_rand = torch.rand_like(x_data, dtype=torch.float) # overrides the datatype of x_data
print(f"Random Tensor: \n {x_rand} \n")
```

```python
# Tensor  attribute describe their datatype,shape,device

tensor = torch.rand(3,4)

print(f"Shape of tensor: {tensor.shape}")
print(f"Datatype of tensor: {tensor.dtype}")
print(f"Device tensor is stored on: {tensor.device}")
```

# **Operation on tensors**

```python
# We move our tensor to the GPU if available
if torch.cuda.is_available():
  tensor = tensor.to('cuda')
```

```python
# Standard numpy like indexing and slicing

tensor = torch.ones(4,4)
print("first row:", tensor[0])
print('First column: ', tensor[:, 0])
print('Last column:', tensor[..., -1])
tensor[:,1] = 0
print(tensor)
```

```python
# joining tensor
# you can use torch.cat to concatenate a sequence of tensors along a given dimension.

t1 = torch.cat([tensor, tensor, tensor], dim=1)
print(t1)
```

```python
# Airthmetic operations.
#By popular demand, the function torch. matmul performs matrix multiplications if both arguments are 2D and 
# computes their dot product if both arguments are 1D . For inputs of such dimensions, its behaviour is the same as np.

#Returns a tensor with the same size as input that is filled with random numbers from a uniform distribution on the interval [ 0 , 1 ) [0, 1) [0,1). 
#torch. rand_like(input) is equivalent to torch.
```

```python
# This computes the matrix multiplication between two tensors. y1, y2, y3 will have the same value
y1 = tensor @ tensor.T
y2 = tensor.matmul(tensor.T)

y3 = torch.rand_like(tensor)
torch.matmul(tensor, tensor.T, out=y3)


# This computes the element-wise product. z1, z2, z3 will have the same value
z1 = tensor * tensor
z2 = tensor.mul(tensor)

z3 = torch.rand_like(tensor)
torch.mul(tensor, tensor, out=z3)
```

```python
#Single element tensor
#If you have a one-element tensor, for example by aggregating all values of a tensor into one value, you can convert it to a Python numerical value using item()
```

```python
agg = tensor.sum()
agg_item = agg.item()  
print(agg_item, type(agg_item))
```

```python
#In-place operations
#Operations that store the result into the operand are called in-place. They are denoted by a _ suffix. For example: x.copy_(y), x.t_(), will change x.
#Note: In-place operations save some memory, but can be problematic when computing derivatives because of an immediate loss of history. Hence, their use is discouraged.
```

```python
print(tensor, "\n")
tensor.add_(5)
print(tensor)
```

```python
# Tensor to nUmpy array

t = torch.ones(5)
print(f"t: {t}")
n = t.numpy()
print(f"n: {n}")
```

```python
#A change in the tensor reflects in the NumPy array.
```

```python
t.add_(1)
print(f"t: {t}")
print(f"n: {n}")
```

```python
# Numpy array to tensor
n = np.ones(5)
t = torch.from_numpy(n)
```

```python
# Changes in the NumPy array reflects in the tensor.
```

```python
np.add(n, 1, out=n)
print(f"t: {t}")
print(f"n: {n}")
```
